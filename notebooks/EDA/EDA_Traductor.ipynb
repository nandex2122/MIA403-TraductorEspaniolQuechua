### Parte 1: Instalación e Importación de Librerías y Descarga de NLTK

```python
# %% [markdown]
# ## 4. EDA Rápido del Corpus
#
# ### 4.1. Instalación e Importación de Librerías

# %%
# Instalar librerías necesarias si aún no las tienes
# !pip install pandas matplotlib seaborn wordcloud nltk numpy

# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
import numpy as np # Para calcular percentiles y unificar resultados

# Descargar recursos de NLTK (solo la primera vez que los uses)
# 'stopwords' es necesario para filtrar palabras comunes en español
# 'punkt' es necesario para tokenizar palabras
try:
    nltk.data.find('corpora/stopwords')
except nltk.downloader.DownloadError:
    print("Descargando stopwords de NLTK...")
    nltk.download('stopwords')
    print("Descargando punkt de NLTK...")
    nltk.download('punkt')
    print("Descargas de NLTK completadas.")

# %%
```

### Parte 2: Carga y Parsing del Corpus

```python
# %% [markdown]
# ### 4.2. Carga y Parsing del Corpus

# %%
# Definir la ruta de tu archivo de corpus combinado
# *******************************************************************
# *** ¡MUY IMPORTANTE! AJUSTA ESTA RUTA A LA UBICACIÓN DE TU ARCHIVO ***
# *******************************************************************
corpus_file = 'corpus_total_formateado.txt' # Ejemplo: Si está en la misma carpeta que el notebook
# O una ruta completa si es diferente:
# corpus_file = 'D:/Jose/UNI/Maestria IA/Cursos/03_Tercer_ciclo/03_Proyecto_Tesis/Proyecto/Diccionarios/huggingface/somosnlp-hackathon-2022/Dataset/corpus_combinado.txt'

# Lista para almacenar los pares de oraciones
data = []

# Patrón de expresión regular para extraer el código, español y quechua
# Asumiendo el formato "CÓDIGO ESPAÑOL CÓDIGO QUECHUA"
patron = re.compile(r"^\d{6}\s(.+?)\s\d{6}\s(.+)$")

try:
    with open(corpus_file, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f):
            line = line.strip()
            if not line: # Ignorar líneas vacías
                continue
            
            match = patron.match(line)
            if match:
                esp_text = match.group(1).strip()
                que_text = match.group(2).strip()
                data.append({'id': i, 'es': esp_text, 'qu': que_text})
            else:
                print(f"Advertencia: Línea {i+1} no coincide con el patrón y será ignorada: {line}")
    
    df_corpus = pd.DataFrame(data)
    print(f"Corpus cargado exitosamente. Total de {len(df_corpus)} pares de oraciones.")

    if df_corpus.empty:
        print("El DataFrame está vacío. Por favor, verifica el archivo y el patrón de extracción.")
    else:
        print("\nPrimeras 5 filas del DataFrame:")
        print(df_corpus.head())

except FileNotFoundError:
    print(f"Error: El archivo '{corpus_file}' no fue encontrado. Por favor, verifica la ruta.")
except Exception as e:
    print(f"Ocurrió un error al cargar el corpus: {e}")

# %%
```

### Parte 3: Hallazgos Principales (Análisis Numérico)

```python
# %% [markdown]
# ### 4.3. Hallazgos Principales (Análisis Numérico)

# %%
if not df_corpus.empty:
    # 1. Número total de pares de oraciones
    total_pares = len(df_corpus)
    print(f"\n1. Número total de pares de oraciones en el corpus: {total_pares}")

    # 2. Longitud de oraciones (en número de palabras)
    df_corpus['es_len'] = df_corpus['es'].apply(lambda x: len(x.split()))
    df_corpus['qu_len'] = df_corpus['qu'].apply(lambda x: len(x.split()))

    print("\n2. Estadísticas de longitud de oraciones (en palabras):")
    print("   Español:")
    print(df_corpus['es_len'].describe())
    print(f"   Percentil 95 (Español): {np.percentile(df_corpus['es_len'], 95):.0f} palabras")
    print("\n   Quechua:")
    print(df_corpus['qu_len'].describe())
    print(f"   Percentil 95 (Quechua): {np.percentile(df_corpus['qu_len'], 95):.0f} palabras")

    # 3. Identificación de oraciones vacías o muy cortas (después del parsing)
    empty_es = df_corpus[df_corpus['es_len'] == 0]
    empty_qu = df_corpus[df_corpus['qu_len'] == 0]

    print(f"\n3. Oraciones vacías o con 0 palabras (después del parsing):")
    print(f"   - Español: {len(empty_es)} ({len(empty_es)/total_pares:.2%})")
    print(f"   - Quechua: {len(empty_qu)} ({len(empty_qu)/total_pares:.2%})")

    # 4. Palabras más comunes (excluyendo stopwords)
    stop_words_es = set(stopwords.words('spanish'))
    
    # Para Quechua, NLTK no tiene stopwords predefinidas.
    # Si tienes una lista de stopwords en Quechua, puedes cargarla aquí:
    # Por ejemplo: stop_words_qu = set(['kay', 'icha', 'ñuqanchik', ...])
    # Para este ejemplo, no filtraremos stopwords específicas para Quechua si no tienes una lista.
    
    all_words_es = ' '.join(df_corpus['es']).lower()
    # Filtra palabras que no son alfabéticas (números, puntuación) y stopwords de español
    words_es = [word for word in nltk.word_tokenize(all_words_es) if word.isalpha() and word not in stop_words_es]
    freq_dist_es = nltk.FreqDist(words_es)
    
    all_words_qu = ' '.join(df_corpus['qu']).lower()
    # Para Quechua, solo filtra palabras que no son alfabéticas
    words_qu = [word for word in nltk.word_tokenize(all_words_qu) if word.isalpha()]
    freq_dist_qu = nltk.FreqDist(words_qu)

    print("\n4. Las 10 palabras más comunes (excluyendo stopwords en español):")
    print("   Español:", freq_dist_es.most_common(10))
    print("   Quechua:", freq_dist_qu.most_common(10)) # Puede que veas palabras comunes si no hay stopwords específicas
else:
    print("No se puede realizar el análisis numérico: el DataFrame del corpus está vacío.")

# %%
```

### Parte 4: Visualizaciones

```python
# %% [markdown]
# ### 4.4. Visualizaciones Clave

# %%
if not df_corpus.empty:
    # Configuración de los gráficos
    plt.style.use('seaborn-v0_8-darkgrid') # Estilo más moderno para los gráficos
    plt.figure(figsize=(15, 6))

    # Histograma de Longitud de Oraciones (Español)
    plt.subplot(1, 2, 1) # 1 fila, 2 columnas, primer gráfico
    sns.histplot(df_corpus['es_len'], bins=30, kde=True, color='skyblue')
    plt.title('Distribución de Longitud de Oraciones (Español)')
    plt.xlabel('Número de Palabras')
    plt.ylabel('Frecuencia')
    plt.grid(axis='y', alpha=0.75)

    # Histograma de Longitud de Oraciones (Quechua)
    plt.subplot(1, 2, 2) # 1 fila, 2 columnas, segundo gráfico
    sns.histplot(df_corpus['qu_len'], bins=30, kde=True, color='lightcoral')
    plt.title('Distribución de Longitud de Oraciones (Quechua)')
    plt.xlabel('Número de Palabras')
    plt.ylabel('Frecuencia')
    plt.grid(axis='y', alpha=0.75)

    plt.tight_layout() # Ajusta el diseño para que no se superpongan los títulos/etiquetas
    plt.show()

    # Nubes de Palabras
    # Une todas las palabras limpias para generar la nube
    # Asegúrate de que las cadenas de texto no estén vacías
    text_es = ' '.join(words_es)
    text_qu = ' '.join(words_qu)

    if text_es:
        wordcloud_es = WordCloud(width=800, height=400, background_color='white').generate(text_es)
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud_es, interpolation='bilinear')
        plt.axis('off') # Ocultar ejes
        plt.title('Nube de Palabras Comunes (Español)')
        plt.show()
    else:
        print("No hay suficientes palabras en español para generar la nube de palabras.")

    if text_qu:
        wordcloud_qu = WordCloud(width=800, height=400, background_color='white').generate(text_qu)
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud_qu, interpolation='bilinear')
        plt.axis('off') # Ocultar ejes
        plt.title('Nube de Palabras Comunes (Quechua)')
        plt.show()
    else:
        print("No hay suficientes palabras en quechua para generar la nube de palabras.")
else:
    print("No se pueden generar visualizaciones: el DataFrame del corpus está vacío.")

# %%
```
