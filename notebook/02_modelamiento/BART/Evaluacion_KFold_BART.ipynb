```python
import pandas as pd
from datasets import Dataset
from tqdm import tqdm  # Para barra de progreso

def load_and_prepare_data(csv_path):
    # Leer el CSV especificando las columnas correctas
    try:
        df = pd.read_csv(
            csv_path,
            usecols=['source', 'target'],  # Solo cargar las columnas necesarias
            encoding='utf-8',
            on_bad_lines='warn'  # Mostrar advertencias si hay líneas mal formateadas
        )
    except Exception as e:
        print(f"Error al leer el CSV: {e}")
        return Dataset.from_list([])
    
    # Verificar datos faltantes
    print("\nResumen de datos faltantes:")
    print(df.isnull().sum())
    
    # Limpiar y preparar los datos
    data_pairs = []
    for _, row in tqdm(df.iterrows(), total=len(df), desc="Procesando filas"):
        if pd.notna(row['source']) and pd.notna(row['target']):
            data_pairs.append({
                "spanish": str(row['source']).strip(),  # Español (source)
                "quechua": str(row['target']).strip()   # Quechua (target)
            })
    
    print(f"\nTotal de pares válidos cargados: {len(data_pairs)}")
    
    # Estadísticas básicas
    if data_pairs:
        avg_len_spanish = sum(len(pair['spanish']) for pair in data_pairs) / len(data_pairs)
        avg_len_quechua = sum(len(pair['quechua']) for pair in data_pairs) / len(data_pairs)
        print(f"\nLongitud promedio:")
        print(f"- Español: {avg_len_spanish:.1f} caracteres")
        print(f"- Quechua: {avg_len_quechua:.1f} caracteres")
    
    return Dataset.from_list(data_pairs)

# Ruta al archivo CSV
csv_file_path = r"C:\DATA\FERNANDOHC\EDUCACION\MAESTRIA\UNI_MAI\SEMESTRE_3\MIA-204ProyectoDeInvestigacion1\Proyecto_Traductor_Esp_Quechua\datos\corpus_trad.csv"

# Cargar datos
raw_dataset = load_and_prepare_data(csv_file_path)

# Verificación detallada
if len(raw_dataset) > 0:
    print("\nEjemplos cargados (formato español -> quechua):")
    for i in range(min(3, len(raw_dataset))):
        print(f"\nEjemplo {i+1}:")
        print(f"Español: {raw_dataset[i]['spanish']}")
        print(f"Quechua: {raw_dataset[i]['quechua']}")
        
    # Verificación de distribución
    sample = raw_dataset.shuffle().select(range(5))
    print("\nMuestra aleatoria:")
    for example in sample:
        print(f"\nEspañol: {example['spanish']}")
        print(f"Quechua: {example['quechua']}")
else:
    print("\nNo se cargaron datos válidos. Verifica:")
    print("1. Que las columnas se llamen 'source' y 'target'")
    print("2. Que el archivo contenga datos")
    print("3. Que no haya caracteres especiales corruptos")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from datasets import Dataset
import torch
from nltk.translate.bleu_score import corpus_bleu
from rouge_score import rouge_scorer
import numpy as np
import pandas as pd

# --- Configuración ---
CSV_PATH = r"C:\DATA\FERNANDOHC\EDUCACION\MAESTRIA\UNI_MAI\SEMESTRE_3\MIA-204ProyectoDeInvestigacion1\Proyecto_Traductor_Esp_Quechua\datos\corpus_trad.csv"
MODEL_PATH_V1 = r"C:\DATA\FERNANDOHC\EDUCACION\MAESTRIA\UNI_MAI\SEMESTRE_3\MIA-204ProyectoDeInvestigacion1\Proyecto_Traductor_Esp_Quechua\modelos\BART\v1"
MODEL_PATH_V2 = r"C:\DATA\FERNANDOHC\EDUCACION\MAESTRIA\UNI_MAI\SEMESTRE_3\MIA-204ProyectoDeInvestigacion1\Proyecto_Traductor_Esp_Quechua\modelos\BART\v2"
NUM_SENTENCES = 50  # Número de oraciones para evaluación rápida
NUM_RUNS = 5       # Número de ejecuciones para calcular media y desviación

# --- Cargar y preparar el dataset ---
df = pd.read_csv(CSV_PATH)
raw_dataset = Dataset.from_dict({
    "spanish": df["source"].tolist(),  # español es la fuente
    "quechua": df["target"].tolist()   # quechua es el objetivo
})

# Verificar GPU
device = "cuda" if torch.cuda.is_available() else "cpu"

# --- Función para calcular métricas ---
def calculate_metrics(predictions, references):
    # BLEU Score
    bleu = corpus_bleu([[ref.split()] for ref in references], 
                      [pred.split() for pred in predictions])
    
    # ROUGE-L Score
    scorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
    rouge_scores = [scorer.score(ref, pred)['rougeL'].fmeasure 
                   for ref, pred in zip(references, predictions)]
    rouge = np.mean(rouge_scores)
    
    return {"bleu": bleu, "rouge": rouge}

# --- Función para evaluar un modelo ---
def evaluate_model(model_path, dataset_subset):
    # Cargar modelo y tokenizer
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_path).to(device)
    
    predictions = []
    references = []

    for example in dataset_subset:
        # Tokenizar entrada (ahora español es la fuente)
        inputs = tokenizer(example["spanish"], return_tensors="pt", 
                         max_length=128, truncation=True, padding="max_length").to(device)
        
        # Generar predicción (quechua es el objetivo)
        output = model.generate(**inputs, max_length=128)
        pred_text = tokenizer.decode(output[0], skip_special_tokens=True)
        
        predictions.append(pred_text)
        references.append(example["quechua"])  # Comparar con el quechua real
    
    return calculate_metrics(predictions, references)

# --- Evaluación múltiple para obtener estadísticas ---
bleu_scores_v1 = []
rouge_scores_v1 = []
bleu_scores_v2 = []
rouge_scores_v2 = []

for run in range(NUM_RUNS):
    print(f"\n--- Ejecución {run+1}/{NUM_RUNS} ---")
    
    # Seleccionar subconjunto aleatorio diferente en cada ejecución
    small_dataset = raw_dataset.shuffle(seed=run).select(range(NUM_SENTENCES))
    
    # Evaluar modelo v1 (baseline)
    metrics_v1 = evaluate_model(MODEL_PATH_V1, small_dataset)
    bleu_scores_v1.append(metrics_v1["bleu"])
    rouge_scores_v1.append(metrics_v1["rouge"])
    print(f"v1 - BLEU: {metrics_v1['bleu']:.4f}, ROUGE-L: {metrics_v1['rouge']:.4f}")
    
    # Evaluar modelo v2 (optimizado)
    metrics_v2 = evaluate_model(MODEL_PATH_V2, small_dataset)
    bleu_scores_v2.append(metrics_v2["bleu"])
    rouge_scores_v2.append(metrics_v2["rouge"])
    print(f"v2 - BLEU: {metrics_v2['bleu']:.4f}, ROUGE-L: {metrics_v2['rouge']:.4f}")

# --- Calcular estadísticas ---
# Medias
mean_bleu_v1 = np.mean(bleu_scores_v1)
mean_bleu_v2 = np.mean(bleu_scores_v2)
mean_rouge_v1 = np.mean(rouge_scores_v1)
mean_rouge_v2 = np.mean(rouge_scores_v2)

# Desviaciones estándar
std_bleu_v1 = np.std(bleu_scores_v1)
std_bleu_v2 = np.std(bleu_scores_v2)
std_rouge_v1 = np.std(rouge_scores_v1)
std_rouge_v2 = np.std(rouge_scores_v2)

# Diferencias
delta_bleu = mean_bleu_v2 - mean_bleu_v1
delta_rouge = mean_rouge_v2 - mean_rouge_v1

# --- Mostrar resultados comparativos ---
print("\n=== RESULTADOS COMPARATIVOS ===")
# BLEU
print(f"\nBLEU Baseline (v1): {mean_bleu_v1:.4f} ± {std_bleu_v1:.4f}")
print(f"BLEU Refinado (v2): {mean_bleu_v2:.4f} ± {std_bleu_v2:.4f}")
print(f"Δ BLEU: {delta_bleu:+.4f}")

# ROUGE-L
print(f"\nROUGE-L Baseline (v1): {mean_rouge_v1:.4f} ± {std_rouge_v1:.4f}")
print(f"ROUGE-L Refinado (v2): {mean_rouge_v2:.4f} ± {std_rouge_v2:.4f}")
print(f"Δ ROUGE-L: {delta_rouge:+.4f}")

# --- Mostrar ejemplos de la última ejecución ---
print("\n=== EJEMPLOS DE TRADUCCIÓN ===")
print("(De la última ejecución aleatoria)")

tokenizer_v1 = AutoTokenizer.from_pretrained(MODEL_PATH_V1)
tokenizer_v2 = AutoTokenizer.from_pretrained(MODEL_PATH_V2)
model_v1 = AutoModelForSeq2SeqLM.from_pretrained(MODEL_PATH_V1).to(device)
model_v2 = AutoModelForSeq2SeqLM.from_pretrained(MODEL_PATH_V2).to(device)

for i in range(min(3, NUM_SENTENCES)):
    # Obtener ejemplo
    example = small_dataset[i]
    
    # Generar predicciones (español -> quechua)
    inputs_v1 = tokenizer_v1(example["spanish"], return_tensors="pt", 
                           max_length=128, truncation=True, padding="max_length").to(device)
    output_v1 = model_v1.generate(**inputs_v1, max_length=128)
    pred_v1 = tokenizer_v1.decode(output_v1[0], skip_special_tokens=True)
    
    inputs_v2 = tokenizer_v2(example["spanish"], return_tensors="pt", 
                           max_length=128, truncation=True, padding="max_length").to(device)
    output_v2 = model_v2.generate(**inputs_v2, max_length=128)
    pred_v2 = tokenizer_v2.decode(output_v2[0], skip_special_tokens=True)
    
    # Mostrar resultados
    print(f"\nEjemplo {i+1}:")
    print(f"Español (Fuente): {example['spanish']}")
    print(f"Quechua (Referencia): {example['quechua']}")
    print(f"v1 (Baseline): {pred_v1}")
    print(f"v2 (Optimizado): {pred_v2}")
```

```python
# --- Mostrar resultados comparativos en formato de tabla ---
print("\n=== RESULTADOS COMPARATIVOS POR FOLD ===")
print("| Fold | BLEU Baseline | BLEU Refinado | Δ BLEU | ROUGE-L Baseline | ROUGE-L Refinado | Δ ROUGE-L |")
print("|---|---|---|---|---|---|---|")

for run in range(NUM_RUNS):
    print(f"| {run+1} | {bleu_scores_v1[run]:.2f} | {bleu_scores_v2[run]:.2f} | {bleu_scores_v2[run]-bleu_scores_v1[run]:+.2f} | "
          f"{rouge_scores_v1[run]:.2f} | {rouge_scores_v2[run]:.2f} | {rouge_scores_v2[run]-rouge_scores_v1[run]:+.2f} |")

# --- Mostrar medias y desviaciones estándar ---
print("\n| Media ±STD | "
      f"{mean_bleu_v1:.2f} ±{std_bleu_v1:.3f} | "
      f"{mean_bleu_v2:.2f} ±{std_bleu_v2:.3f} | "
      f"{delta_bleu:+.2f} | "
      f"{mean_rouge_v1:.2f} ±{std_rouge_v1:.3f} | "
      f"{mean_rouge_v2:.2f} ±{std_rouge_v2:.3f} | "
      f"{delta_rouge:+.2f} |")
```
